```python
import requests, pandas as pd
```

```python
# This data is from the world bank.

df=pd.read_csv("https://raw.githubusercontent.com/joesellis/joesellis.github.io/main/Forest%20area%20(%25%20of%20total%20land%20area).csv", skiprows=6, encoding='latin-1')
```

```python
df
# We see that there are rows we don't want at the bottom, so drop all data after Zimbabwe

df = df.iloc[:191]
```

```python
df
```

```python
# There are some empty columns.

df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df
```

```python
df.columns
```

```python
# I am only interested in a few of these countries, so I create a list.

countries = [' Brazil', ' China', ' Indonesia', ' Togo', ' Dominican Republic', ' United Kingdom', ' United States', ' China', 'Australia', ' Viet Nam', ' Uruguay', ' Germany']
df2 = df.loc[df['Country'].isin(countries)]

df2
```

```python
# The format of the data currently won't be interactive, so I modify it using the melt function

df3 = df2.melt(id_vars=['Country'], 
value_vars=['1990','1995','2000','2005','2010','2011','2012','2013','2014','2015','2016'], var_name='Year', 
value_name='Forest Coverage')
```
