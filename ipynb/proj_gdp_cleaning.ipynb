```python
import requests, pandas as pd

# Bringing in the dataframe.
df=pd.read_csv("https://raw.githubusercontent.com/joesellis/joesellis.github.io/main/DP_LIVE_31122021124810224.csv", encoding='latin-1')

df
```

```python
# Renaming relevant columns

df.columns = ['Country Code', '.', '..','...','....','Quarter', 'GDP Growth', '.....']
```

```python
df
```

```python
# I am only interested in plotting a few countries.

codes = [
          "CHN",
          "GBR",
          "OECD",
          "USA",
          "IND",
          "RUS",
          "JPN",
          "BRA",
          "DEU",
          "ARG"
        ]
df = df.loc[df['Country Code'].isin(codes)]
```

```python
df
```

```python
# I am only interested in the country, time, and GDP growth.

df2 = df[['Country Code', 'Quarter', 'GDP Growth']]
```

```python
df2
```

```python
# I want to add country name information, it currently isn't in the OECD dataset. This function adds a full country name to my dataframe.

def Country_Name (row):
   if row['Country Code'] == 'CHN' :
      return 'China'
   if row['Country Code'] == 'GBR' :
      return 'United Kingdom'
   if row['Country Code'] == 'OECD' :
      return 'OECD'
   if row['Country Code'] == 'USA':
      return 'United States'
   if row['Country Code']  == 'IND':
      return 'India'
   if row['Country Code'] == 'RUS':
      return 'Russia'
   if row['Country Code'] == 'JPN':
      return 'Japan'
   if row['Country Code']  == 'BRA':
      return 'Brazil'
   if row['Country Code'] == 'DEU':
      return 'Germany'
   if row['Country Code'] == 'ARG':
      return 'Argentina'

   # In case of error   
   return 'other'
```

```python
df2['Country'] = df2.apply (lambda row: Country_Name(row), axis=1)
```

```python
df2
```

```python
# I want the full date as well as the quarter, it makes for a better plot. Using a timestamp function.

qs = df['Quarter']
qs

df2['Date'] = pd.PeriodIndex(qs, freq='Q').to_timestamp()
df2
 
```

```python
# Exporting

df2.to_csv("proj_gdp_clean.csv",encoding = 'utf-8', index=False)
```
